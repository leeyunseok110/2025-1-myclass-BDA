## 시리즈 활용

```python
import numpy as np
import pandas as pd

s = ['희빈', '수영', '현수', '지호', '지민']
p = pd.Series([99, 84, 92, 65, 78], index=s)
p
```

```python
p.name = '수학성적'
p.index.name = '이름'
p
```

```python
p.sort_index()
```

```python
p.sort_index(ascending=False)
```

```python
p.sort_values()
```

```python
p.sort_values(ascending=False)
```

```python
print(p.max())
print(p.min())
print(p.mean())
print(p.median())
print(p.count())
print(p.size)
```

```python
p.describe()
```

```python
s = pd.Series([3,1,1,2,None,np.nan,4,6,6,np.nan])
s
```
