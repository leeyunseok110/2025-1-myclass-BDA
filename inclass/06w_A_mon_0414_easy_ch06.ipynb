```python
df2 = pd.DataFrame([[1, 3000], [5, 500]],
                  columns = ['id', 'price'])
df2
```

```python
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, on = 'id', how='inner')
```

```python
df1.merge( df2, on = 'id', how='inner')
```
