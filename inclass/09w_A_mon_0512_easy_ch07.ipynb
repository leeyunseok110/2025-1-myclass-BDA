```python
p25,p50,p75 = mpg.hwy.quantile([.25,.5,.75])
p25,p50,p75
```

```python
iqr = p75 - p25
iqr
```

```python
#upper fence
UF = p75 + 1.5 * iqr
UF
```

```python
#lower fence
LF = p25 - 1.5 * iqr
LF
```

```python
mpg.hwy.describe()
```

```python
mpg.hwy.describe()['50%']
```

```python
mpg[ mpg.hwy > UF ][['hwy']]
```
